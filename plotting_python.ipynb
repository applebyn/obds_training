#### Matplotlib

- Basic plotting package 
- Very customisable - every single line/pixel 
- Need alot of code to write a simple plot

```python
#an 'axes' is a frame/pane within a larger figure/window.
#an axis is x-axis or y-axis



# Make x & y variables using numpy 
x = np.arange(0, 5, 0.1)
y = np.sin(x)

print('This is x:')
print(x)

print()

print('This is y:')
print(y)
```

### Matplot lib object orientated interface

```python

fig = plt.figure()   # create figure object - contain axes, graphics, text
ax = plt.axes()      # create axes - a single pane

ax.plot(x, y) # create continuous line plot

ax.set_xlim(0,10)
ax.set_ylim(-1.5,1.5)
ax.set_title('title of plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

### Matplot lib Pyplot interface

```python
plt.plot(x, y)
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
plt.title('title of plt')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

#### Scatterplot

#### Pairs plot

```python
# A pairs plot combines different types of data

iris = sns.load_dataset("iris")
display(iris.head())

# plot pairs plot
# assign plot to object 'g' - sets up grid
g = sns.PairGrid(iris, hue="species")

# plot histogram on diagonal
g.map_diag(plt.hist) 
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend();
```

### Lineplot example

```python
fmri = sns.load_dataset("fmri")
display(fmri.head())
sns.relplot(x="timepoint",
            y="signal", 
            hue="event", 
            kind="line", 
            data=fmri)

sns.relplot(x="timepoint", 
            y="signal", 
            hue="region", 
            style="event",
            dashes=False, 
            markers=True, 
            kind="line", 
            data=fmri);
```

### Facets

```python
sns.relplot(x="timepoint",
            y="signal",
            hue="subject",
            col="region",
            row="event",
            height=3,
            kind="line",
            estimator=None,
            data=fmri)
```

Normalise read counts to total reads per sample total reads & convert to 'counts per million (CPM)'


```python
#sum the columns for each sample
sum_counts = filtered_counts_gene_index.sum()
display(filtered_counts_gene_index)
display(sum_counts)
# divide the dataframe by the sum of the columns
filtered_counts_gene_index.div(sum_counts, axis = 1)
```

```python
# now, convert that dataframe into counts per million
# divide by the sum of the columns and multiple by 10^6. cpm is a useful normalisation tool
filtered_counts_cpm = filtered_counts_gene_index.div(sum_counts, axis = 1)*1000000
filtered_counts_cpm
```
